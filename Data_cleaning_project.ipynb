```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customer_shopping_data.csv.zip")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
sns.boxplot(x=df['price'])
plt.show()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

IQR = Q3 - Q1

df = df[(df['price'] >= Q1 - 1.5*IQR) & (df['price'] <= Q3 + 1.5*IQR)]
```

```python
sns.boxplot(x=df['price'])
plt.show()
```

```python
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.hist(df['age'])
plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(x='category', data=df)
plt.xticks(rotation=45)
plt.show()
```

Insight 1 :
The age distribution shows that most customers fall within the young to middle-age group. This indicates that people in this age range are more actively involved in shopping activities.

Insight 2 :
The gender distribution suggests that both male and female customers participate actively in shopping, indicating balanced purchasing behavior across genders.

Insight 3 :
The category analysis reveals that certain product categories receive significantly higher purchases than others, indicating strong customer preferences toward these categories.

Insight 4 :
The purchase amount distribution indicates that most customers prefer products within the mid-price range, suggesting a preference for moderately priced items.

Insight 5 :
Overall analysis suggests that businesses should focus their marketing strategies on popular product categories and target the most active age groups to increase customer engagement and sales.
